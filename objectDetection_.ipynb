Importing libraries


```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```




Reading the Image



```python

img_path = "/content/drive/MyDrive/Lab1/BasPic.jpg"

img_bgr = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)


img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)


plt.imshow(img_rgb)
plt.title("Original Image (RGB)")
plt.axis("off")
plt.show()


plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

Print Image Dimensions

```python
h, w = img_rgb.shape[:2]
print("Image Height =", h)
print("Image Width  =", w)
```

Pixel Values at (10,10), (50,100), (120,201)




```python
coords = [(10,10), (50,100), (120,201)]

for r,c in coords:
    print(f"Pixel at ({r},{c}) - RGB:", img_rgb[r,c])
    print(f"Pixel at ({r},{c}) - Gray:", img_gray[r,c])
```

Extract bold textROI (Height: 50-200, Width: 10-40)

```python

y1, y2 = 5, 300
x1, x2 = 350, 650

roi = img_rgb[y1:y2, x1:x2]

plt.figure(figsize=(4,4))
plt.imshow(roi)
plt.title(f"Face ROI (Rows {y1}-{y2}, Cols {x1}-{x2})")
plt.axis("off")
plt.show()

```

Rotate Image Clockwise (90)

This rotates the image into right direction.



```python
rot_cw = cv2.rotate(img_rgb, cv2.ROTATE_90_CLOCKWISE)

plt.figure(figsize=(5,5))
plt.imshow(rot_cw)
plt.title("Rotation: 90 Clockwise")
plt.axis("off")
plt.show()
```

Rotate Image Anti-Clockwise (90)

This rotates the image left direction.

```python
rot_acw = cv2.rotate(img_rgb, cv2.ROTATE_90_COUNTERCLOCKWISE)

plt.figure(figsize=(5,5))
plt.imshow(rot_acw)
plt.title("Rotation: 90 Anti-Clockwise")
plt.axis("off")
plt.show()
```

Rotate Image by 180
This flips the image upside-down

```python
rot_180 = cv2.rotate(img_rgb, cv2.ROTATE_180)

plt.figure(figsize=(5,5))
plt.imshow(rot_180)
plt.title("Rotation: 180")
plt.axis("off")
plt.show()
```

Horizontal Flip (value = 1)

```python
flip_horizontal = cv2.flip(img_rgb, 1)

plt.figure(figsize=(5,5))
plt.imshow(flip_horizontal)
plt.title("Flip: Horizontal (LeftRight)")
plt.axis("off")
plt.show()
```

Vertical Flip (value = 0)

```python
flip_vertical = cv2.flip(img_rgb, 0)

plt.figure(figsize=(5,5))
plt.imshow(flip_vertical)
plt.title("Flip: Vertical (UpDown)")
plt.axis("off")
plt.show()
```

Flip Both Axes (value = -1)

```python
flip_both = cv2.flip(img_rgb, -1)

plt.figure(figsize=(5,5))
plt.imshow(flip_both)
plt.title("Flip: Both Axes")
plt.axis("off")
plt.show()
```

Draw Rectangle


```python
img_rect = img_rgb.copy()


cv2.rectangle(img_rect, (x1, y1), (x2, y2), (255, 0, 0), 5)


cv2.putText(img_rect, "Basgul Nikzad - 212068",
            (20, h - 30), cv2.FONT_HERSHEY_SIMPLEX,
            1, (0, 0, 255), 3)

plt.figure(figsize=(6,6))
plt.imshow(img_rect)
plt.title("Basgul Nikzad - 212068 ")
plt.axis("off")
plt.show()

```

Add 10% Border Around Image

```python

top = bottom = int(0.10 * h)
left = right = int(0.10 * w)

bordered_img = cv2.copyMakeBorder(
    img_rgb, top, bottom, left, right,
    cv2.BORDER_CONSTANT, value=(0,0,0)
)

plt.figure(figsize=(6,6))
plt.imshow(bordered_img)
plt.title("Image with 10% Black Border")
plt.axis("off")
plt.show()
```

500500 White Image with 4 Colored Boxes

```python

white_img = np.ones((500,500,3), dtype=np.uint8) * 255

box_size = 500 // 8

white_img[0:box_size, 0:box_size] = [255, 0, 0]

white_img[0:box_size, 500 - box_size:500] = [0, 255, 0]

white_img[500 - box_size:500, 0:box_size] = [0, 0, 255]

white_img[500 - box_size:500, 500 - box_size:500] = [0, 0, 0]

plt.figure(figsize=(5,5))
plt.imshow(white_img)
plt.title("500x500 Image with 4 Colored Boxes")
plt.axis("off")
plt.show()
```
